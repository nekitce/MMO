```python
import numpy as np
df["genres"].unique()
genres = df["genres"].values
print(genres[1])
allgenres = []
for i in genres:
  str = i.split()
  allgenres = np.concatenate([allgenres, str])
allgenres
```

```python
allgenresdf = pd.DataFrame(data = allgenres)
allgenresdf
```

```python
uniqueGenres = allgenresdf[0].unique()
d = dict.fromkeys(uniqueGenres, 0)
for i in genres:
  str = i.split()
  for i in str:
    d[i] = d[i] + 1
d
```
